7. Given a number, check if it's positive, and if it is also even. If not positive, print if it's zero or negative.

```python
num =  int(input("Enter a number:"))
print("Entered number:",num)
if num > 0:
    if num % 2 == 0:
        print("Number is positive Even")
    else:
        print("Number is Positive Odd")
elif num == 0:
    print("Number is Zero")
else:
    print("Number is Negative")
```

11. Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
* Use .info(), .describe()
* Select 'Name' and 'City' columns
* Drop 'City' column
* Fill any missing values in 'Salary' column with the mean
* Remove any duplicate rows

```python
import pandas as pd
```

```python
data = {
    "Name":["Philbert","Sakshi","Omkar","Samyak","Aman","Aneesh","Omkar","Rohan","Pratham","Aman"],
    "Age":[25,21,26,23,25,28,26,25,29,30],
    "City":["Belgaum","Bangalore","Mumbai","Bangalore","Delhi","Mumbai","Mumbai","Delhi","Mumbai","Pune"],
    "Salary":[50000,45000,60000,None,72000,59000,60000,None,40000,50000]
}
```

```python
df_data = pd.DataFrame(data)
df_data
```

```python
df_data.info()
```

```python
df_data.describe()
```

```python
df_data[["Name","City"]]
```

```python
df = df_data.drop("City",axis=1)
df
```

```python
df["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df
```

```python
df.drop_duplicates()
```
