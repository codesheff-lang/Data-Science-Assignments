# ASSIGNMENT 06(B): HYPOTHESIS TESTING

```python
import numpy as np
from scipy import stats
```

```python
print("Step 1: Hypothesis")
print("H0: μ = 4000 (mean weekly cost follows the model)")
print("H0: μ > 4000 (mean weekly cost is higher than the model)")
```

```python
x_mean = 3050
tht_mean = 1000 + 5 * 600
sigma_x = 25
sigma = 5 * sigma_x
n = 25
alpha = 0.05
```

```python
standard_error = sigma / np.sqrt(n)
z = (x_mean - tht_mean) / standard_error
print("\n Step 2 : Test Statistics")
print(f"Theoretical Mean : {tht_mean}")
print(f"Population Std Dev: {sigma}")
print(f"standard_error: {standard_error}")
print(f"z-statistics: {z:.4f}")
```

```python
z_critical = stats.norm.ppf(1 - alpha)
p_value = 1 - stats.norm.cdf(z)
print("\n Step 3: Probability and Comparison")
print(f"Critical z-value (alpha=0.05): {z_critical:.4f}")
print(f"p-value: {p_value:.4f}")
print(f"Compare: p-value > alpha? {p_value > alpha}")
```

```python
if z > z_critical:
    decision = "Reject H0"
else:
    decision = "Do not reject H0"

print("Step 4: Decision")
print(f"Decision: {decision} (since z = {z:.4f} < {z_critical:.4f})")
```

```python
print("Step 5: Conclusion")
print("There is not strong evidence that weekly operating costs are higher than the model suggests.")
```
