Exercise 1: Prime Numbers

Prime nummber are those which can't be divided by any other number, which can only divided by itself and 1

```python
n = int(input("Enter the number:"))

if n>1 and all(n % i != 0 for i in range(2, int(n**0.5)+1)):
    print(n, "is a prime number")
else:
    print(n, "is not a prime number")
```

Excercise 2: Product of two random numbers

Develop a Python program that generates two random numbers and asks the user to enter the product of these numbers. The program should then check if the user's answer is correct and display an appropriate message.

```python
import random

num1 = random.randint(1,10)
num2 = random.randint(1,10)

correct_product = num1*num2

user_answer = int(input(f"What is the product of {num1} and {num2}?"))

if user_answer == correct_product:
    print("Correct! Well Done.")
else:
    print(f"Incorrect. The correct answer is {correct_product}.")
```

Exercise 4: Hospital Billing System


✅ Problem Statement (Question): Hospital Billing System A hospital charges its patients based on the following: • Room charges per day: ₹2,000 • Doctor consultation fee (flat): ₹1,500 • Lab test charges: Based on the number of tests taken. Each test costs ₹300 • Medicine charges: Total cost of medicines provided • Discount: If the total bill (before discount) exceeds ₹10,000, a 10% discount is applied Write a Python program to calculate the final bill for a patient who: • Stayed for 4 days • Had 3 lab tests • Had medicine charges of ₹2,400

```python
room_charge_per_day = 2000
doctor_fee = 1500
cost_per_test = 300
days_stayed = 4
lab_tests = 3
medicine_charge = 2400

room_charge = days_stayed * room_charge_per_day
lab_charge = lab_tests * cost_per_test

total_before_discount = room_charge + doctor_fee + lab_charge + medicine_charge

discount = 0
if total_before_discount > 10000:
    discount = total_before_discount * 0.10

final_bill = total_before_discount - discount 
print(f"Room Charges ( {days_stayed} days x {room_charge_per_day}):{room_charge}")
print(f"Doctor Consultation fee: {doctor_fee}")
print(f"lab_tests ({lab_tests} x {cost_per_test}): {lab_charge}")
print(f"medicine_charge: {medicine_charge}")
print(f"Total_before_discount: {total_before_discount}")
if discount > 0:
    print(f"Discount (10%): {discount}")
print(f"Final Bill Amount: {final_bill}")
```
